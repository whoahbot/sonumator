```python
import pandas as pd
import glob

path = "output/csv"
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    cols = pd.read_csv(filename, nrows=1).columns
    df = pd.read_csv(filename, index_col=None, usecols=[cols[1]])
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
